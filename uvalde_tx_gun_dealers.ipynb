```python
import pandas as pd
```

```python
df = pd.read_csv('data/uvalde_gun_dealers.csv')
```

df.columns

```python
cols = []
for i in df.columns:
    cols.append(i.lower())
df.columns = cols
```

```python
df['address'] = df['premise_street'] + ', ' + df['premise_city'] + ', ' + df['premise_state'] + ' ' + df['premise_zip_code']
```

```python
df = df.rename(columns = {'latitude': 'lat', 'longitude':'long'})
df.columns
```

```python
print(len(df))
df.drop_duplicates(subset =['license_name', 'lic_type', 'address'], keep='first', inplace = True)
len(df)
```

```python
# 01  Dealer in Firearms Other Than Destructive Devices (Includes Gunsmiths)
# 02  Pawnbroker in Firearms Other Than Destructive Devices
# 03  Collector of Curios and Relics
# 06  Manufacturer of Ammunition for Firearms
# 07  Manufacturer of Firearms Other Than Destructive Devices
# 08  Importer of Firearms Other Than Destructive Devices
# 09  Dealer in Destructive Devices
# 10  Manufacturer of Destructive Devices
# 11  Importer of Destructive Devices

df2 = df.loc[((df['lic_type'] == 1) | (df['lic_type'] == 2))].groupby('premise_state', as_index=False).count().sort_values('license_name', ascending=False)
df2 = df2[['premise_state', 'license_name']]
df2 = df2.rename(columns = {'premise_state': 'state', 'license_name':'count'})
df2 = df2.reset_index()
df2.drop('index', axis=1, inplace=True)
df2
```

```python
df2['median'] = df2['count'].median()
df2['mean'] = df2['count'].mean()

df2['times_median'] = df2['count'] / df2['median']
df2['times_mean'] = df2['count'] / df2['mean']
```

```python
abbrevs = pd.read_csv('data/state_abbrevs.csv')

df2 = pd.merge(df2, abbrevs, on='state', how='left')

df2
```

```python
pop = pd.read_csv('data/census_state_pop.csv')
df2 = pd.merge(df2, pop, on='state_name', how='inner')
df2
```

```python
df2['dealers_per_cap'] = df2['count'] / df2['Population Estimate, July 1, 2021']
df2['dealers_per_100k'] = df2['dealers_per_cap'] * 100000
df2.sort_values('dealers_per_100k', ascending=False).reset_index()
```
