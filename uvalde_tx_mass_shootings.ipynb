```python
import pandas as pd
```

```python
df = pd.read_csv('data/everytown_massshootings.csv')
df
```

```python
df2 = df.groupby('State', as_index=False).count()
df2 = df2[['State', 'Date']]
df2 = df2.rename(columns = {'Date': 'count'})
df2 = df2.sort_values('count', ascending=False)
df2 = df2.reset_index()
df2
```

```python
df.groupby('State', as_index=False).sum()
```

```python
df.groupby('State', as_index=False).sum().sort_values('Total shot', ascending=False)
```

```python
df3 = df.groupby('State', as_index=False).sum()
df3 = df3[['State', 'Total shot', 'Number killed', 'Number wounded']]
df3 = df3.rename(columns = {'Total shot':'total_shot', 'Number killed':'total_killed', 'Number wounded': 'total_wounded'})
df2 = pd.merge(df2, df3, on='State', how='inner')
df2.drop('index', axis=1, inplace=True)
```

```python
abbrevs = pd.read_csv('data/state_abbrevs.csv')

not_in = []

for i in abbrevs['state']:
    if i not in list(df2['State']):
        not_in.append(i)
        
notin = abbrevs.loc[abbrevs['state'].isin(not_in)]
notin = notin.rename(columns = {'state': 'State'})
notin.drop('state_name', axis=1, inplace=True)
notin['count'] = 0
notin['total_shot'] = 0
notin['total_killed'] = 0
notin['total_wounded'] = 0

df2 = pd.concat([df2, notin], ignore_index=True)

df2
```

```python
df2 = df2.sort_values('total_killed', ascending=False).reset_index()
df2.drop('index', axis=1, inplace=True)
df2
```

```python
df2.to_csv('data/everytown_massshootings_sum.csv', index=False)
```
